```python
X=ml_df.iloc[:,:-1]
y=ml_df.iloc[:,-1]
```

```python
X
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit_transform(X_train)
scaler.transform(X_test)
```

### Linear regression model ###

```python
from sklearn.linear_model import  LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
reg.coef_
reg.intercept_
```

```python
y_pred=reg.predict(X_test)
```

```python
plt.scatter(y_pred,y_test)
```

```python
from sklearn.metrics import  r2_score,mean_squared_error,root_mean_squared_error,mean_absolute_error
linear_score1=r2_score(y_pred,y_test)
linear_score2=mean_squared_error(y_test,y_pred)
linear_score3=root_mean_squared_error(y_test,y_pred)
linear_score4=mean_absolute_error(y_test,y_pred)
linear_score1
```

```python
reg.coef_
```

#### Ridge regression model #####

```python
from sklearn.linear_model import Ridge 
reg2=Ridge()
reg2.fit(X_train,y_train)
y_pred_1=reg2.predict(X_test)
```

```python
y_pred_1
```

```python
ridge_score1=r2_score(y_pred_1,y_test)
ridge_score2=mean_squared_error(y_test,y_pred_1)
ridge_score3=root_mean_squared_error(y_test,y_pred_1)
ridge_score4=mean_absolute_error(y_test,y_pred_1)
ridge_score1
```

```python
reg2.coef_
```

### Lasso regression model ###

```python
from sklearn.linear_model import Lasso
```

```python
lasso_reg=Lasso()
lasso_reg.fit(X_train,y_train)
y_pred_lasso=lasso_reg.predict(X_test)
```

```python
lasso_score1=r2_score(y_pred_lasso,y_test)
lasso_score2=mean_squared_error(y_test,y_pred_lasso)
lasso_score3=root_mean_squared_error(y_test,y_pred_lasso)
lasso_score4=mean_absolute_error(y_test,y_pred_lasso)
lasso_score1
```

```python
lasso_reg.coef_
```

### Elasticnet regression model ###

```python
from sklearn.linear_model import ElasticNet
EN_reg=ElasticNet()
EN_reg.fit(X_train,y_train)
y_pred_EN=EN_reg.predict(X_test)
```

```python
EN_score1=r2_score(y_pred_EN,y_test)
EN_score2=mean_squared_error(y_test,y_pred_EN)
EN_score3=root_mean_squared_error(y_test,y_pred_EN)
EN_score4=mean_absolute_error(y_test,y_pred_EN)
EN_score1
```

```python
EN_reg.coef_
```
